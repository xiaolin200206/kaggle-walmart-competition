```python

import pandas as pd
import numpy as np
import lightgbm as lgb
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
```

```python
# Load data
train = pd.read_csv("D:/xmapp/competition/walmart_train.csv")
test = pd.read_csv("D:/xmapp/competition/walmart_test.csv")
feature = pd.read_csv("D:/xmapp/competition/walmart_feature.csv")
store = pd.read_csv("D:/xmapp/competition/walmart_store.csv")
```

```python
#print the each columns
print("Train Columns:", train.columns)
print("Test Columns:", test.columns)
print("Feature Columns:", feature.columns)
print("Store Columns:", store.columns)

# Ensure column names are correct
feature.rename(columns=lambda x: x.strip(), inplace=True)  # Remove spaces
feature.rename(columns={"date": "Date"}, inplace=True)  # Fix lowercase issue

# Convert "Date" column to datetime format
feature["Date"] = pd.to_datetime(feature["Date"], errors="coerce")
train["Date"] = pd.to_datetime(train["Date"], errors="coerce")
test["Date"] = pd.to_datetime(test["Date"], errors="coerce")

# Merge datasets
train = train.merge(feature, on=["Store", "Date"], how="left")
train = train.merge(store, on="Store", how="left")

test = test.merge(feature, on=["Store", "Date"], how="left")
test = test.merge(store, on="Store", how="left")

# Remove duplicate columns
train = train.loc[:, ~train.columns.duplicated()]
test = test.loc[:, ~test.columns.duplicated()]

# Print first few rows to check
print(train.head())
print(test.head())
```

```python
#feature engieering and handling missing value
train['Date'] = pd.to_datetime(train['Date'])
test['Date'] = pd.to_datetime(test['Date'])
train["Year"] = train["Date"].dt.year
train["Month"] = train["Date"].dt.month
train["Week"] = train["Date"].dt.isocalendar().week

test["Year"] = test["Date"].dt.year
test["Month"] = test["Date"].dt.month
test["Week"] = test["Date"].dt.isocalendar().week
train.drop(columns=["Date"], inplace=True)
test.drop(columns=["Date"], inplace=True)
```

```python
X = train.drop(columns = ["Weekly_Sales","IsHoliday_x"])
Y = train["Weekly_Sales"]

#why use validation??
#validation is used to fine-tune hyperparameter and detect overfitting

#why don't use validation??
#in a small dataset, splitting further into a validation set reduces training data
# a better approach for small dataset is cross validation

#when to use each method:
#use train-validation-test split when large dataset
#use k-fold cross-validation when small dataset
#use a validation set or cross-validation when we prepare to hyperparameter tuning
#use the test set only once after tuning wehn we final model evaluation
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,random_state=42)

#convert categorical columns
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

x_train,X_test = X_train.align(X_test, join="left", axis = 1)
```

```python
#n_estimators means the number of boosting iterations or trees in your ensemble
#it define how many trees will be built sequentially, each tree is added to correct the errors made by the previous trees
#a high number of estimators allow the model to capture more complex patterns

#max_depth mean the maximum depth of each individual tree
#limits how deep each tree can grow
#a deeper tree can model more complex relationships but may overfit the training data
#max_depth=6 restrict trees to a moderate level of complexity
#it help in preventing overfitting, ensure data do not become too tailored to the training data
#the optimal depth depends on your data complexity and noise levels
#xgb_model = xgb.XGBRegressor(n_estimators = 100, learning_rate = 0.1, max_depth=5, random_state=42)
#xgb_model.fit(X_train,Y_train)

#what's pipeline??
#pipeline in mechine learning is a sequence of data processing steps that automate workflows
#ensure consistency and effiency. it is used to streamline the process of data transform, model
#training, and prediction

#why use pipeline??
#eliminates manual execution of multiple steps
#ensure consistent results across different runs 
#reduce redundant operations,especially in feature engineering
#ensure that transformation are only applied to training data before model fitting

#when to use a pipeline??
#preprocessing is required--scalling, encoding, or handing missing value
#multiple step need to be executed in sequenc--feature transformation\
#prevent data leakage--ensure data preprocessing is only applied to the training set before fitting the model
#improve model effiency and maintainability -- reduce redundant code and ensure consistency
#hyperparameter tuning-- when using gridsearch cv or randomizedsearchcv, a pipeline ensure all step and applied during cross validation
pipeline = Pipeline([
    ('xgb_model',XGBRegressor(n_estimators = 100, learning_rate = 0.1, max_depth=5, random_state=42))
])
pipeline.fit(X_train,Y_train)
y_pred = pipeline.predict(X_test)


#how to understand this step??
#use the trained xgboost model:the model parttern has already learnt from the training data
#this step is for make prediction

#make prediction on x_resr: x_test contains only feature
#the model takes x_test as input and generate corresponding prediction y_pred

#store the predicted value in y_pred
#y_pred will contain the model's predicted sale values for each row in x_test
#Y_pred = xgb_model.predict(X_test)
```

```python
#calculate wmae
#how to understand??
#test["Weekly_Sales"] mean extract the actual weekly sales from the test dataset
#reindex(X_test.index) ensure that y_true has the same raw indices as x_test
#why do we need to do this?
#we shuffled the data during the split, x_test contains only a subset of test, and we need to align the indices of
#weekly_sales with test
#y_true will have mismatched rows if we don't do this

#must we do this step before calculate the evaluate metrics??
#of course yes, because must ensure correct data alignment
#Y_true = test["Weekly_Sales"].reindex(X_test.index) get error, so why use Y_test instend of test["weekly_sales"]

Y_true = Y_test.reindex(X_test.index)
weight = test["IsHoliday_x"].reindex(X_test.index).apply(lambda x:5 if x else 1)

def wmae(Y_true, weight, Y_pred):
    return np.sum(weight * np.abs(Y_true - Y_pred) / np.sum(weight))
print(wmae(Y_true, weight, Y_pred))
```

```python
#what's residual??
#residual is the difference between the actual value and the predict value
#small residual mean prediction are close to actual value which is good model
#large residual mean prediction are far from actual value which is bad model

#when should we analyze residual
#before model deployment to check if the model is reliable
#after model training to identify improvement areas
#if model accuracy is low to find why it's making errore
residuals = Y_test - Y_pred
sns.histplot(residuals, bins=50, kde=True)
plt.title("Distribution of Prediction Errors")
plt.xlabel("Residual Error")
plt.show()

#how to understand this output??
#most residual are center around 0 ,means this model is generally accurate
#a sharp peak at 0 suggests that many prediction are close to actual sale

#potential issues:
#the long right tail suggests that some prediction severely underestimate sales
#if the model consistently under-predict, it might not be capturing high-sales trend properly
#there might be missing feature or a need for better feature engineering
```
